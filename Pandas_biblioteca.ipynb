## **Demandas de ML**

### Cuales son los valores promedio de alquiler por tipo de propiedad?

```python
datos.head()
```

```python
# Este es el valor promedio de la columna "Valor"
datos["Valor"].mean()
```

```python
# Con esta funcion sacamos el promedio de la columna "valor" para cada "tipo" y los ordenamos por "valor"
# Y almacenamos esta consulta en una variable
df_tipo_precio=datos.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
df_tipo_precio
```

## **Demandas de DEV**

## **Crear columnas numericas (operaciones entre columnas)**

#### 1. valor_mensual: Esta columna debe contener los gastos mensuales de cada propiedad, incluyendo el alquiler y el condominio

```python
# Esto crea una nueva columna que suma los datos de "Valor" y "Condominio" para cada fila
datos["Valor mensual"]=datos["Valor"]+datos["Condominio"]
datos.head()
```

#### 2. valor_anual: Esta columna debe contener los gastos anuales por propiedad, es decir, el impuesto mas 12 meses de alquiler y condominio

```python
datos["Valor anual"]=datos["Valor mensual"]*12+datos["Impuesto"]
datos.tail()
```

## **Crear columnas categoricas**

#### Descripcion: esta columna debe contener un resumen de la informacion clave de las propiedades que se mostraran en el sitio web: tipo de propiedad, barrio, cantidad de habitaciones y plazas de estacionamiento

```python
datos["Descripcion"]=datos["Colonia"]+" con "+datos["Habitaciones"].astype(str)+" habitaciones y "+datos["Garages"].astype(str)+" plazas de estacionamiento"
datos.head()
```

#### Tiene_suite: Esta debe ser una columna que indique unicamente si la propiedad tiene o no suites, sin importar la cantidad

```python
datos["Tiene suite"]=datos["Suites"].apply(lambda x: "Si" if x>0 else "No")
datos.head()
```
