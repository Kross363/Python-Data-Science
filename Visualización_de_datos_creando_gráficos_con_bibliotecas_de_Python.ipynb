###**Visualizacion de datos con Python: Matplotlib, Seaborn y Plotly**

```python
import pandas as pd
df=pd.read_csv("inmigrantes_canada.csv")
df.head()
```

```python
df.info()
```

```python
#Indicamos que la columna pais sea el array que cuenta las columnas
df.set_index("Pais",inplace=True)
df.head()
```

```python
anios=list(map(str,range(1980,2014)))
colombia=df.loc["Colombia",anios]
colombia
```

```python
col_dict={"Año":colombia.index.tolist(),
          "Inmigrantes":colombia.values.tolist()}
datos_col=pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
plt.plot(datos_col["Año"],datos_col["Inmigrantes"])
plt.xticks(["1980","1985","1990","1995","2000","2005","2010"])
plt.title("Inmigrantes de Colombianos hacia Canadá")
plt.xlabel("Año")
plt.ylabel("Inmigrantes")
plt.grid(True)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(8,4))
ax.plot(datos_col["Año"],datos_col["Inmigrantes"])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title("Inmigrantes de Colombianos hacia Canadá")
ax.set_xlabel("Año")
ax.set_ylabel("Inmigrantes")
ax.grid(True)
plt.show()
```

```python
fig,axs=plt.subplots(1,2,figsize=(10,4))
axs[0].plot(datos_col["Año"],datos_col["Inmigrantes"])
axs[0].set_title("Inmigracion de Colombianos hacia canadá")
axs[0].set_xlabel("Año")
axs[0].set_ylabel("Inmigrantes")
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid(True)
axs[1].boxplot(datos_col["Inmigrantes"])
axs[1].set_title("Boxplot de la inmigración de Colombianos hacia Canadá")
axs[1].set_xlabel("Colombia")
axs[1].set_ylabel("Numero de inmigrantes")
axs[1].grid(True)

plt.show()
```

```python
fig,axs=plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle("Comparación entre países")

#Ver en que posicion de axs[x,y] coloca cada grafico
axs[0,0].plot(df.loc["Colombia",anios])
axs[0,0].set_title("Colombia")
axs[0,1].plot(df.loc["Brasil",anios])
axs[0,1].set_title("Brasil")
axs[1,0].plot(df.loc["Argentina",anios])
axs[1,0].set_title("Argentina")
axs[1,1].plot(df.loc["Perú",anios])
axs[1,1].set_title("Perú")

ymin=0
ymax=7000

#Con este for le colocamos las caracteristicas a todos los graficos

for ax in axs.ravel():
  ax.set_ylim([ymin,ymax])
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel("Año")
  ax.set_ylabel("Inmigrantes")
  ax.grid(True)
plt.show()
```
