###**Visualizacion de datos con Python: Matplotlib, Seaborn y Plotly**

```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/inmigrantes_canada.csv")
df.head()
```

```python
df.info()
```

```python
#Indicamos que la columna pais sea el array que cuenta las columnas
df.set_index("Pais",inplace=True)
df.head()
```

```python
anios=list(map(str,range(1980,2014)))
colombia=df.loc["Colombia",anios]
colombia
```

```python
col_dict={"Año":colombia.index.tolist(),
          "Inmigrantes":colombia.values.tolist()}
datos_col=pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
plt.plot(datos_col["Año"],datos_col["Inmigrantes"])
plt.xticks(["1980","1985","1990","1995","2000","2005","2010"])
plt.title("Inmigrantes de Colombianos hacia Canadá")
plt.xlabel("Año")
plt.ylabel("Inmigrantes")
plt.grid(True)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(8,4))
ax.plot(datos_col["Año"],datos_col["Inmigrantes"])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title("Inmigrantes de Colombianos hacia Canadá")
ax.set_xlabel("Año")
ax.set_ylabel("Inmigrantes")
ax.grid(True)
plt.show()
```

```python
fig,axs=plt.subplots(1,2,figsize=(10,4))
axs[0].plot(datos_col["Año"],datos_col["Inmigrantes"])
axs[0].set_title("Inmigracion de Colombianos hacia canadá")
axs[0].set_xlabel("Año")
axs[0].set_ylabel("Inmigrantes")
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid(True)
axs[1].boxplot(datos_col["Inmigrantes"])
axs[1].set_title("Boxplot de la inmigración de Colombianos hacia Canadá")
axs[1].set_xlabel("Colombia")
axs[1].set_ylabel("Numero de inmigrantes")
axs[1].grid(True)

plt.show()
```

```python
datos_col.describe()
```

```python
fig,axs=plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle("Comparación entre países")

#Ver en que posicion de axs[x,y] coloca cada grafico
axs[0,0].plot(df.loc["Colombia",anios])
axs[0,0].set_title("Colombia")
axs[0,1].plot(df.loc["Brasil",anios])
axs[0,1].set_title("Brasil")
axs[1,0].plot(df.loc["Argentina",anios])
axs[1,0].set_title("Argentina")
axs[1,1].plot(df.loc["Perú",anios])
axs[1,1].set_title("Perú")

ymin=0
ymax=7000

#Con este for le colocamos las caracteristicas a todos los graficos

for ax in axs.ravel():
  ax.set_ylim([ymin,ymax])
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel("Año")
  ax.set_ylabel("Inmigrantes")
  ax.grid(True)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(8,3))
ax.plot(datos_col["Año"],datos_col["Inmigrantes"],lw=2.5,color="g")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title("Inmigrantes de Colombianos hacia Canadá",fontsize=18,loc="left")
ax.set_xlabel("Año",fontsize=14)
ax.set_ylabel("Inmigrantes",fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
ax.grid(linestyle="--")
plt.show()
```

```python
sudamerica=df.query('Region=="América del Sur"')
sudamerica.head(10)
```

```python
colores=["royalblue","orange","forestgreen","orchid","purple","brown","slateblue","gray","olive","navy","teal","tomato"]
fig,ax=plt.subplots(figsize=(12,5))
ax.bar(sudamerica.index,sudamerica["Total"],color=colores)
ax.set_title("Inmigracion de sudamericanos a Canadá",loc="left",fontsize=18)
ax.set_xlabel("")
ax.set_ylabel("Numero de inmigrantes",fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
colores=["royalblue","orange","forestgreen","orchid","purple","brown","slateblue","gray","olive","navy","teal","tomato"]
fig,ax=plt.subplots(figsize=(12,5))
ax.barh(sudamerica.index,sudamerica["Total"],color=colores)
ax.set_title("Inmigracion de sudamericanos a Canadá",loc="left",fontsize=18)
ax.set_ylabel("")
ax.set_xlabel("Numero de inmigrantes",fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
sudamerica_sorted=sudamerica.sort_values("Total",ascending=True)
colores=["royalblue","orange","forestgreen","orchid","purple","brown","slateblue","gray","olive","navy","teal","tomato"]
fig,ax=plt.subplots(figsize=(12,5))
ax.barh(sudamerica_sorted.index,sudamerica_sorted["Total"],color=colores)
ax.set_title("Inmigracion de sudamericanos a Canadá desde 1980 hasta 2013",loc="left",fontsize=18)
ax.set_ylabel("")
ax.set_xlabel("Numero de inmigrantes",fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
colores=[]
for pais in sudamerica_sorted.index:
  if pais=="Colombia":
    colores.append("forestgreen")
  else:
    colores.append("silver")
fig,ax=plt.subplots(figsize=(12,5))
ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.barh(sudamerica_sorted.index,sudamerica_sorted["Total"],color=colores)
ax.set_title("Sudamerica: Colombia es el segundo pais con mas migrantes hacia Canadá desde 1980 hasta 2013",loc="left",fontsize=18)
ax.set_ylabel("")
ax.set_xlabel("Numero de inmigrantes",fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
for i,j in enumerate(sudamerica_sorted["Total"]):
  ax.text(j+20,i,str(j),color="black",fontsize=8.5,ha="left",va="center")

#Exportando el grafico
fig.savefig("/content/drive/MyDrive/colombia_a_canada.png",dpi=300,bbox_inches="tight",transparent=False)
plt.show()
```
