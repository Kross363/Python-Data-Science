### Importación de datos



```python
import pandas as pd
from matplotlib import pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_total=tienda['Precio'].sum()
print(f"El ingreso total de la Tienda N°1 es de: ${ingreso_total}")
```

```python
ingreso_total=tienda2['Precio'].sum()
print(f"El ingreso total de la Tienda N°2 es de: ${ingreso_total}")
```

```python
ingreso_total=tienda3['Precio'].sum()
print(f"El ingreso total de la Tienda N°3 es de: ${ingreso_total}")
```

```python
ingreso_total=tienda4['Precio'].sum()
print(f"El ingreso total de la Tienda N°4 es de: ${ingreso_total}")
```

# 2. Ventas por categoría

```python
conteo_por_categoria = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)

print("--- Conteo de Productos Vendidos por Categoría (Tienda 1) ---")
plt.pie(conteo_por_categoria, labels=conteo_por_categoria.index, autopct='%1.1f%%')
plt.title('Conteo de Productos Vendidos por Categoría')
plt.show()
print(conteo_por_categoria)
```

```python
conteo_por_categoria = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)

print("--- Conteo de Productos Vendidos por Categoría (Tienda 2) ---")
plt.bar(conteo_por_categoria.index, conteo_por_categoria.values)
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Ventas')
plt.title('Conteo de Productos Vendidos por Categoría')
plt.show()
print(conteo_por_categoria)
```

```python
conteo_por_categoria = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)

print("--- Conteo de Productos Vendidos por Categoría (Tienda 3) ---")
print(conteo_por_categoria)
```

```python
conteo_por_categoria = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)

print("--- Conteo de Productos Vendidos por Categoría (Tienda 4) ---")
plt.pie(conteo_por_categoria, labels=conteo_por_categoria.index, autopct='%1.1f%%')
plt.title('Conteo de Productos Vendidos por Categoría')
plt.show()
print(conteo_por_categoria)
```

# 4. Productos más y menos vendidos

```python
conteo_por_producto = tienda.groupby('Categoría del Producto').size()

# Identificar el Producto Más Vendido
producto_mas_vendido = conteo_por_producto.idxmax() # Obtiene el índice (nombre del producto) con el valor máximo
ventas_max = conteo_por_producto.max()             # Obtiene el valor máximo de ventas

# Identificar el Producto Menos Vendido
producto_menos_vendido = conteo_por_producto.idxmin() # Obtiene el índice (nombre del producto) con el valor mínimo
ventas_min = conteo_por_producto.min()             # Obtiene el valor mínimo de ventas

# Mostrar los resultados
print("---  Producto Más Vendido (Tienda 1) ---")
print(f"Producto: **{producto_mas_vendido}**")
print(f"Total de Ventas: **{ventas_max}**")

print("\n---  Producto Menos Vendido (Tienda 1) ---")
print(f"Producto: **{producto_menos_vendido}**")
print(f"Total de Ventas: **{ventas_min}**")
```
